## **مقدمه**

در حوزه‌ی **بازیابی اطلاعات (IR)**، کیفیت و کارایی نتایج تا حد زیادی به نحوه‌ی آماده‌سازی داده‌های متنی وابسته است.  
متون خام معمولاً شامل نویسه‌های نامعتبر، تفاوت در نگارش کلمات، واژه‌های زائد و ساختارهای زبانی پیچیده هستند که بدون پردازش مناسب می‌توانند عملکرد سیستم را تضعیف کنند.  

بنابراین، پیش‌پردازش داده‌ها به‌عنوان **یکی از مهم‌ترین مراحل در چرخه‌ی بازیابی اطلاعات** شناخته می‌شود. این فرآیند مجموعه‌ای از عملیات است که داده‌ی متنی را از حالت خام به وضعیتی ساخت‌یافته و قابل‌تحلیل برای مدل‌های جستجو، شاخص‌گذاری (Indexing) و تحلیل معنایی تبدیل می‌کند.

در ادامه، گام‌های اصلی این فرآیند به‌صورت دقیق و تحلیلی توضیح داده می‌شوند.



## **۱. پالایش نویسه‌های غیرحرفی (Non-alphabetic Character Filtering)**

### **توضیح فنی:**
در مرحله‌ی ابتدایی، متن خام از نظر وجود کاراکترهایی که ماهیت زبانی ندارند، پالایش می‌شود.  
به کمک **عبارات منظم (Regular Expressions - Regex)** تمام نویسه‌هایی که جزو حروف الفبایی نیستند — از جمله اعداد، نشانه‌های نگارشی، علائم ریاضی، کاراکترهای خاص یا فاصله‌های تکراری — شناسایی و حذف می‌شوند.

### **اهمیت در IR:**
این مرحله برای جلوگیری از ورود نویز به فضای برداری (Vector Space) ضروری است. وجود نمادها و کاراکترهای غیرحرفی می‌تواند باعث افزایش ابعاد غیرواقعی در شاخص‌گذاری و کاهش دقت شباهت متنی شود.

### **اثر تحلیلی:**
پس از اجرای این گام، مجموعه‌داده فقط شامل واژه‌های معنادار باقی می‌ماند که این امر به **افزایش دقت در وزن‌دهی واژه‌ها (Term Weighting)** و **بهبود شاخص‌سازی (Index Construction)** منجر می‌شود.


## **۲. قطعه‌بندی واژگانی (Tokenization)**

### **توضیح فنی:**
در این مرحله، هر سند متنی به **واحدهای زبانی کوچک‌تر** تقسیم می‌شود که به آن‌ها **توکن (Token)** گفته می‌شود.  
توکن‌ها معمولاً شامل کلمات منفرد هستند، اما در برخی سیستم‌ها ممکن است شامل عبارات چندکلمه‌ای یا اصطلاحات تخصصی نیز باشند. برای انجام این فرآیند از روش‌های مبتنی بر فضاها (Whitespace Tokenization) یا الگوریتم‌های پیشرفته‌تر نظیر **NLTK Tokenizer** استفاده می‌شود.

### **اهمیت در IR:**
توکن‌سازی هسته‌ی اصلی تمام فرآیندهای بعدی در بازیابی اطلاعات است. بدون آن، هیچ ساختار قابل‌اتکایی برای تحلیل فراوانی، شباهت، و نمایه‌سازی ایجاد نخواهد شد.

### **اثر تحلیلی:**
تقسیم متن به توکن‌ها، مسیر را برای محاسبه‌ی **فراوانی واژه‌ها (Term Frequency - TF)** و محاسبه‌ی وزن‌های **TF-IDF** هموار می‌سازد و در نهایت منجر به افزایش دقت جستجو و شباهت میان اسناد می‌شود.


## **۳. نرمال‌سازی حروف (Case Normalization)**

### **توضیح فنی:**
در زبان انگلیسی و بسیاری از زبان‌های دیگر، وجود حروف بزرگ و کوچک می‌تواند باعث ایجاد دو نمایه‌ی متفاوت از یک واژه شود. برای جلوگیری از این پدیده، تمام حروف به حالت کوچک (**Lowercase**) تبدیل می‌شوند.  
به عنوان مثال، کلمات “Computer” و “computer” پس از این مرحله به شکل یکسانی ثبت می‌شوند.

### **اهمیت در IR:**
اگر نرمال‌سازی انجام نشود، واژه‌های یکسان در فهرست واژگان (Vocabulary) به صورت مجزا ذخیره می‌شوند که منجر به **افزایش ابعاد غیرضروری در شاخص** و **کاهش دقت رتبه‌بندی (Ranking)** خواهد شد.

### **اثر تحلیلی:**
این گام سبب یکنواخت‌سازی نمایش واژه‌ها و افزایش سازگاری در محاسبات آماری می‌شود. در نتیجه، تحلیل شباهت بین اسناد دقیق‌تر و منسجم‌تر صورت می‌گیرد.


## **۴. حذف واژه‌های توقفی (Stop Words Removal)**

### **توضیح فنی:**
در متون طبیعی، واژه‌هایی وجود دارند که از نظر معنا نقش محتوایی چندانی ندارند و تقریباً در تمامی اسناد تکرار می‌شوند؛ مانند “and”، “in”، “the”، “of” و نظایر آن.  
در این مرحله، مجموعه‌ای از این واژگان (Stop Word List) تعریف شده و از متن حذف می‌شوند.

### **اهمیت در IR:**
وجود این کلمات نه‌تنها به تمایز مفهومی اسناد کمکی نمی‌کند، بلکه باعث افزایش حجم داده و زمان پردازش در هنگام **Indexing** می‌شود.  
با حذف آن‌ها، تمرکز سیستم روی واژگان محتوایی و کلیدی‌تر افزایش می‌یابد.

### **اثر تحلیلی:**
این عملیات باعث افزایش نسبت سیگنال به نویز (Signal-to-Noise Ratio) در داده‌ها می‌شود و در نتیجه کیفیت شاخص‌ها و نتایج جستجو بهبود پیدا می‌کند.


## **۵. ریشه‌یابی واژگان (Stemming)**

### **توضیح فنی:**
در زبان انگلیسی، واژه‌ها معمولاً با پسوندها و پیشوندهای مختلفی ظاهر می‌شوند.  
برای مثال، “connect”، “connected”، “connecting” و “connection” همگی به مفهومی واحد اشاره دارند. در مرحله‌ی ریشه‌یابی، الگوریتمی مانند **Porter Stemmer** این واژه‌ها را به ریشه‌ی اصلی (“connect”) بازمی‌گرداند.

### **اهمیت در IR:**
ریشه‌یابی باعث کاهش حجم واژگان در پایگاه داده و افزایش پوشش معنایی مدل می‌شود. در نتیجه، اگر کاربر واژه‌ای را جستجو کند که شکل صرفی آن با واژه‌های موجود متفاوت باشد، سیستم همچنان قادر به بازیابی اسناد مرتبط خواهد بود.

### **اثر تحلیلی:**
با کاهش واژه‌های هم‌معنی و هم‌ریشه به یک نمایه‌ی واحد، **فضای جستجو (Search Space)** بهینه شده و **Recall** سیستم افزایش می‌یابد، بدون آنکه **Precision** به‌صورت چشم‌گیری کاهش پیدا کند.


## **جمع‌بندی بخش پیش‌پردازش داده‌ها**

فرآیند پیش‌پردازش در بازیابی اطلاعات، یک مرحله‌ی ساده‌ی مقدماتی نیست، بلکه زیربنای تمام الگوریتم‌های جستجو و رتبه‌بندی محسوب می‌شود.  
مراحلی مانند **حذف نویسه‌های غیرالفبایی، توکن‌سازی، نرمال‌سازی، حذف واژه‌های توقفی و ریشه‌یابی**، در کنار هم باعث می‌شوند داده‌های متنی خام به شکلی ساخت‌یافته و تحلیلی تبدیل شوند.  

نتیجه‌ی اجرای دقیق این گام‌ها عبارت است از:

#  پیاده‌سازی و رفتار شاخص معکوس (Inverted Index)

**خلاصهٔ گزارش:**  
در این سند، یک طراحی عملی و با جزئیات از پیاده‌سازی شاخص معکوس ارائه شده است. هدف توضیح مؤلفه‌ها، ساختارهای داده، روش‌های افزودن سند و ساخت شاخص، نحوهٔ بازیابی پستینگ‌ها و محاسبهٔ IDF با و بدون هموارسازی (smoothing) است. متن طوری تنظیم شده که برای پروژه‌ها و گزارش‌های درسی در درس «بازیابی اطلاعات (IR)» قابل‌استفاده و مرجع باشد.


## ۱. نگاه کلی و هدف طراحی

شاخص معکوس ساختاری است که برای نگهداری نقشهٔ «ترم → لیست اسناد‌ حاوی آن ترم» به کار می‌رود. این ساختار زیرساخت اصلی برای عملیات‌هایی مانند محاسبهٔ وزن‌ها (TF, IDF, BM25)، بازیابی اسناد مرتبط و نمایش نتایج جستجو است. پیاده‌سازی مورد نظر به‌گونه‌ای طراحی شده که هم عملیات افزایشی (افزودن سند تکی) و هم ساخت دسته‌ای (batch) را پشتیبانی کند.


## ۲. ساختارهای داده‌ای مورد استفاده

در پیاده‌سازیِ پیشنهادی، چند ساختار کلیدی ذخیره‌سازی به‌کار می‌رود:

- **واژه‌نامهٔ شاخص (index):**  
  ساختاری از نوع دیکشنری که کلیدِ آن‌ها «ترم» (term) است و مقدار هر کلید یک لیست از رکوردهای پستینگ است. هر رکورد پستینگ شامل حداقل دو مقدار می‌باشد: شناسهٔ سند (`doc_id`) و فراوانی ترم در آن سند (`tf`). شکل منطقی:  


index = {

"term1": [(doc_id1, tf1), (doc_id2, tf2), ...],

"term2": [(doc_id3, tf3), ...],

...

}

<div dir="rtl">

- **طول اسناد (doc_lengths):**  
یک دیکشنری که به ازای هر `doc_id`، طول سند (تعداد توکن‌های موجود در آن سند پس از پیش‌پردازش) را ذخیره می‌کند. این مقدار برای محاسبات وزن‌دهی و نرمال‌سازی اهمیت دارد.

- **فهرست توکن‌های هر سند (doc_tokens):**  
نگهداری توکن‌های (کلمات) هر سند به‌صورت جداگانه در صورت نیاز به عملیات بعدی (مانند بازسازی پستینگ یا تحلیل‌های داخلی).

- **تعداد اسناد (doc_count):**  
شمارنده‌ای که تعداد کل اسناد واردشده در شاخص را نگه می‌دارد.

- **تعداد اسنادی که یک ترم در آن‌ها ظاهر شده است (term_doc_freq):**  
یک دیکشنری که برای هر ترم تعداد اسناد حاوی آن ترم (Document Frequency — DF) را نگهداری می‌کند. این مقدار برای محاسبهٔ IDF و وزن‌دهی اهمیتی کلیدی دارد.

- **میانگین طول سند (avg_doc_len):**  
مقدار عددی که میانگین طول اسناد را محاسبه و ذخیره می‌کند؛ برای مدل‌های مبتنی بر طول مثل BM25 کاربرد دارد.



## ۳. رابط‌ها و متدهای اصلی (شرح و منطق داخلی)

در این بخش متدهای اصلی و رفتار هر یک شرح داده می‌شود. نام‌ها نمادین هستند؛ پیاده‌سازی می‌تواند در قالب کلاس با نام `InvertedIndex` صورت گیرد.

### ۳.۱ سازنده / مقداردهی اولیه (`__init__`)
**وظیفه:** آماده‌سازی ساختارهای دادهٔ موردنیاز برای شاخص‌گذاری.  
**عملکرد:** ایجاد دیکشنری‌های خالی برای `index`، `doc_lengths`، `doc_tokens`، `term_doc_freq` و مقداردهی صفر به `doc_count` و `avg_doc_len`.

**ملاحظات طراحی:**  
- می‌توان از ساختارهای بهینه‌شدهٔ حافظه (مثل `defaultdict(list)` یا فرمت‌های فشردهٔ باینری برای نگهداری پستینگ‌ها) استفاده کرد.
- در پیاده‌سازی مقیاس‌پذیر، نگهداری پستینگ‌ها در دیسک یا پایگاه کلیدی-مقداری (KV-store) به‌جای حافظهٔ رم توصیه می‌شود.



### ۳.۲ افزودن یک سند منفرد (`add_document(document)`)
**پارامتر ورودی:** یک شیء یا ساختار که حداقل شامل `doc_id` و `tokens` است (لیست توکن‌های سند).  
**گام‌ها (منطق داخلی):**
1. محاسبهٔ فراوانی واژه‌ها در سند با استفاده از شمارنده (مثلاً `Counter(tokens)`).
2. ذخیرهٔ تعداد توکن‌ها در `doc_lengths[doc_id]`.
3. ذخیرهٔ توکن‌های سند در `doc_tokens[doc_id]`.
4. افزایش شمارندهٔ اسناد `doc_count += 1`.
5. برای هر ترم در شمارندهٔ سند:
 - افزودن یا به‌روزرسانی رکورد پستینگ در `index[term]` به‌صورت افزودن جفت `(doc_id, tf)`.
 - اگر این ترم برای اولین بار در این سند ظاهر شده است (یعنی doc_id جدید)، مقدار `term_doc_freq[term]` را به‌روزرسانی کنید (معمولاً `term_doc_freq[term] += 1`).

**نکتهٔ مهم:** هنگام افزودن تکراری از یک سند با همان `doc_id` باید استراتژی صریحی وجود داشته باشد (مثلاً بازنویسی کامل، ادغام یا نادیده‌گرفتن).

**مزایا:** مناسب برای پایگاه‌هایی که به‌صورت افزایشی (online) اسناد جدید می‌گیرند.


### ۳.۳ ساخت دسته‌ای شاخص (`build_index(documents)`)
**پارامتر ورودی:** مجموعه‌ای از اسناد (لیست یا iterable) که هرکدام شامل `doc_id` و `tokens` هستند.  
**منطق:** این متد رفتار مشابه `add_document` را برای هر سند در مجموعه انجام می‌دهد، اما در حالت دسته‌ای می‌توان بهینه‌سازی‌هایی انجام داد:
- محاسبهٔ `term_doc_freq` و `index` به‌صورت متمرکز (مثلاً ساخت جدول میانی و سپس ادغام) تا از دسترسی‌های متوالی به دیکشنری کاهش یابد.
- پس از پردازش تمام اسناد، محاسبهٔ `avg_doc_len = sum(doc_lengths.values()) / doc_count`.

**مزایا:** ساخت سریع‌تر و کارآمدتر شاخص با امکان بهینه‌سازی حافظه و پردازش (مثلاً استفاده از کماپرسورها یا نوشتن مستقیم به فایل‌های پستینگ).


### ۳.۴ بازیابی پستینگ‌ها (`get_postings(term)`)
**هدف:** برگرداندن لیست پستینگ برای یک ترم درخواست‌شده.  
**خروجی:** یک لیست از جفت‌های `(doc_id, tf)` یا ساختار دیگری که اطلاعات لازم برای رتبه‌دهی را در اختیار قرار می‌دهد (مثلاً همراه با فیلد‌های اضافی مثل موقعیت ترم داخل سند برای پشتیبانی از جستجوی عبارتی).

**عملکرد:**  
- اگر ترم در `index` موجود باشد، پستینگ‌های مرتبط بازگردانده می‌شود؛ در غیر این صورت لیست خالی یا `None` بازگردانده می‌شود.

**بهبودها:**  
- نگهداری پستینگ‌ها به‌صورت مرتب‌شده بر اساس `doc_id` برای اجرای عملیات همپوشانی (intersection) سریع‌تر در پرس‌وجوهای چندترمی.



### ۳.۵ محاسبهٔ شناسهٔ معکوس سند (IDF)

#### ۳.۵.۱ تعریف ریاضی
برای یک ترم معین \( t \):
\[
\text{IDF}(t) \;=\; \log\left(\frac{N}{df(t)}\right)
\]
که در آن \( N \) تعداد کل اسناد (`doc_count`) و \( df(t) \) همان `term_doc_freq[t]` است.

#### ۳.۵.۲ پیاده‌سازی (compute_idf(term))
- اگر \( df(t) > 0 \) محاسبهٔ فوق انجام می‌شود و مقدار عددی برگردانده می‌شود.
- اگر \( df(t) = 0 \) رفتار تابع باید مشخص باشد (معمولاً بازگشت صفر یا مقدار نوسان‌یافته) — در اینجا یک روش مطمئن‌تر استفاده از هموارسازی توصیه می‌شود.


### ۳.۶ محاسبهٔ IDF بدون هموارسازی (`compute_idf_without_smoothing(term)`)
این متد محاسبهٔ IDF را دقیقاً مطابق فرمول بالا انجام می‌دهد و هیچگونه اصلاح یا تنظیم برای جلوگیری از تقسیم بر صفر یا نتایج منفی اعمال نمی‌کند.  
**ملاحظات ایمنی:** قبل از تقسیم باید چک شود که `term_doc_freq[term] != 0` تا از خطای تقسیم بر صفر جلوگیری شود؛ در غیر این صورت باید استراتژی خطا یا مقدار بازگشتی خاص (مثلاً `float('-inf')` یا `0`) تعریف گردد.


## ۴. نکات طراحی و جنبه‌های عملیاتی

- **هماهنگی بین add_document و build_index:**  
- `add_document` برای حالت افزایشی مناسب است؛ `build_index` برای بارگذاری اولیهٔ مجموعهٔ بزرگی از اسناد. پیاده‌سازی باید تضمین کند که هر دو مسیر به یک وضعیت نهایی همگرا می‌شوند.

- **نگهداری مکان‌های واژه (positional postings):**  
- اگر پشتیبانی از پرس‌وجوهای عبارتی (phrase queries) لازم باشد، پستینگ‌ها باید شامل موقعیت‌های ترم داخل سند نیز باشند: `(doc_id, tf, [positions])`.

- **فشرده‌سازی و ذخیره‌سازی:**  
- برای صرفه‌جویی در فضا، معمولاً از فشرده‌سازی دلتا (gap encoding) روی `doc_id`ها و رمزنگاری‌های عددی برای فرکانس‌ها استفاده می‌شود.

- **محاسبات آماری جانبی:**  
- علاوه بر `avg_doc_len`، ممکن است شاخص‌های دیگری مانند انحراف معیار طول اسناد یا توزیع فراوانی واژگان برای تحلیل کیفیت و بهینه‌سازی وزن‌دهی محاسبه شوند.

- **مقابله با اسناد تکراری و به‌روزرسانی شاخص:**  
- اگر اسناد با `doc_id` تکراری اضافه شوند باید سیاستی (overwrite, merge, ignore) مشخص شود؛ به‌روزرسانی‌های مکرر می‌تواند نیاز به بازسازی جزئی یا کامل برخی بخش‌های شاخص داشته باشد.

#  ساختارهای شاخص‌گذاری اسناد در سیستم‌های بازیابی اطلاعات

در سیستم‌های **بازیابی اطلاعات (IR)**، برای انجام جستجوهای سریع و دقیق روی مجموعه‌های بزرگ اسناد، نیاز به ایجاد شاخص‌ها و نمایه‌های مختلف داریم. شاخص‌ها نقش کلیدی در کاهش زمان جستجو، بهبود دقت نتایج و امکان پردازش هوشمند اطلاعات ایفا می‌کنند. در ادامه، انواع شاخص‌های مهم و نقش آن‌ها به صورت دقیق توضیح داده شده است.


## ۱. **واژه‌نامه یا Lexicon**

### **تعریف و ساختار:**
واژه‌نامه (Lexicon) مجموعه‌ای از تمام **کلمات یا ترم‌های منحصربه‌فرد** موجود در مجموعه اسناد است. این ساختار معمولاً به شکل یک لیست یا دیکشنری پیاده‌سازی می‌شود که برای هر کلمه، اطلاعات پایه‌ای مانند شناسه یا آدرس پستینگ‌ها نگهداری می‌شود.

### **هدف استفاده:**
- جلوگیری از پردازش تکراری کلمات در زمان ساخت شاخص.  
- تسهیل جستجوی سریع و مطمئن بین ترم‌ها.  
- فراهم کردن فهرستی از ترم‌های موجود برای الگوریتم‌های وزن‌دهی و مدل‌های زبانی.

### **مثال کاربردی:**
فرض کنید واژه‌های `["apple", "banana", "computer"]` در مجموعه اسناد وجود دارند. Lexicon تضمین می‌کند که هر بار این کلمات شناسایی شده و تنها یکبار پردازش شوند.


## ۲. **شاخص معکوس (Inverted Index یا Posting Index)**

### **تعریف و عملکرد:**
شاخص معکوس، ساختاری است که برای **هر کلمه**، **لیست اسناد حاوی آن کلمه و تعداد تکرار آن در هر سند** را ذخیره می‌کند.  
- هر ورودی در شاخص معکوس معمولاً شامل `(doc_id, term_frequency)` است.  
- به عنوان مثال، برای کلمه‌ی `apple` ممکن است لیست پستینگ به شکل `[(doc1, 2), (doc3, 1)]` باشد.

### **هدف استفاده:**
- امکان **جستجوی سریع بر اساس ترم‌ها** فراهم می‌شود.  
- اسناد مرتبط با یک ترم مشخص بدون نیاز به اسکن کل مجموعه بازیابی می‌شوند.  
- پایه‌ای برای محاسبهٔ وزن‌های TF-IDF و مدل‌های رتبه‌بندی است.


## ۳. **شاخص سند (Document Index)**

### **تعریف و محتوا:**
شاخص سند شامل اطلاعات **مرتبط با هر سند به‌صورت جداگانه** است. اطلاعات معمول شامل:
- شناسه سند (`doc_id`)  
- طول سند (تعداد توکن‌ها)  
- اطلاعات کمکی یا ویژگی‌های اضافی که به مدل‌های بازیابی کمک می‌کند، مانند نوع سند یا برچسب‌های موضوعی.

### **هدف استفاده:**
- دسترسی سریع به مشخصات هر سند برای عملیات رتبه‌بندی و تحلیل‌های جانبی.  
- تسهیل محاسبات وزن‌دهی و نرمال‌سازی طول سند.  
- امکان افزودن داده‌های تکمیلی بدون نیاز به تغییر شاخص معکوس.


## ۴. **شاخص متادیتا (Meta Index)**

### **تعریف و محتوا:**
شاخص متا برای ذخیره اطلاعات **فراتر از محتوای متن** هر سند استفاده می‌شود، مانند:
- نویسنده یا ناشر سند  
- تاریخ انتشار  
- نوع سند (مقاله، خبر، گزارش و غیره)  

### **هدف استفاده:**
- تسهیل **فیلتر کردن نتایج جستجو** بر اساس ویژگی‌های متا.  
- امکان اجرای جستجوهای پیچیده، مانند محدود کردن نتایج به یک نویسنده خاص یا بازهٔ زمانی مشخص.  
- پشتیبانی از تحلیل‌های آماری و گزارش‌دهی برای کاربران و سیستم‌های مدیریتی.

```python
def visualize_bm25_surface(index_data, scorer, query_set, rel_info):
    k_grid = np.arange(0.0, 3.25, 0.25)
    b_grid = np.arange(0.0, 1.25, 0.25)
    k_mesh, b_mesh = np.meshgrid(k_grid, b_grid, indexing='ij')

    def compute_map(k_val, b_val):
        arguments = {'k': round(float(k_val), 2), 'b': round(float(b_val), 2)}
        try:
            results = ev.evaluate_scoring_method(
                index_data, scorer, query_set, rel_info, params=arguments
            )
            return results.get('MAP', 0)
        except Exception:
            return 0

    combos = [
        {'k': k, 'b': b, 'MAP': compute_map(k, b)}
        for k, b in zip(k_mesh.ravel(), b_mesh.ravel())
    ]

    df = pd.DataFrame(combos)
    pivoted = df.pivot(index='k', columns='b', values='MAP').sort_index(ascending=True)
    heat_matrix = pivoted.values

    fig, axis = plt.subplots(figsize=(10, 8))
    image = axis.imshow(heat_matrix, cmap='viridis', aspect='auto')

    axis.set_xticks(np.arange(len(pivoted.columns)))
    axis.set_yticks(np.arange(len(pivoted.index)))

    axis.set_xticklabels([f"{val:.2f}" for val in pivoted.columns])
    axis.set_yticklabels([f"{val:.2f}" for val in pivoted.index])

    plt.setp(axis.get_xticklabels(), rotation=45, ha='right', rotation_mode='anchor')

    for row_idx, k_val in enumerate(pivoted.index):
        for col_idx, b_val in enumerate(pivoted.columns):
            val = heat_matrix[row_idx, col_idx]
            axis.text(col_idx, row_idx, f"{val:.3f}", ha="center", va="center", color="w")

    axis.set_title("BM25 MAP Scores Across k and b")
    axis.set_xlabel("b values")
    axis.set_ylabel("k values")

    fig.colorbar(image, ax=axis, label='MAP')
    plt.tight_layout()
    plt.show()
```

```python
visualize_bm25_surface(
    index_data=inverted_index,
    scorer=rm.bm25_score,
    query_set=queries,
    rel_info=relevance_judgments
)
```

```python
def optimize_parameters(model_name, scoring_function, queries, relevance_info, index_data):
    bounds = {'k': (0.0, 3.0), 'b': (0.0, 1.0), 'sigma': (0.0, 2.0)}
    grid_values = {'k': [0, 1, 2, 3], 'b': [0, 1], 'sigma': [0, 1, 2]}
    model_keys = {
        'BM25': ['k', 'b'],
        'Method 1': [],
        'Method 2': ['k'],
        'Method 3': ['k'],
        'Method 4': [],
        'Method 5': ['k', 'b'],
        'Method 6': ['k', 'b', 'sigma'],
        'Method 7': ['k'],
        'Method 8': ['k'],
        'Pivoted Length V1': ['b'],
        'Pivoted Length V2': ['b']
    }

    step = 0.25
    print(f"\nRunning optimization for model: {model_name}")
    params_in_use = model_keys.get(model_name, [])
    if not params_in_use:
        print("No tunable parameters for this model.")
        return pd.DataFrame([{'MAP': 0}])

    def run_grid(param_grid):
        all_rows = []
        for values in product(*param_grid.values()):
            trial = dict(zip(param_grid.keys(), values))
            try:
                # Only pass parameters that the specific scoring function accepts
                import inspect
                sig = inspect.signature(scoring_function)
                valid_params = {k: v for k, v in trial.items() if k in sig.parameters}
                
                output = ev.evaluate_scoring_method(index_data, scoring_function, queries, relevance_info, params=valid_params)
                trial['MAP'] = output.get('MAP', 0)
            except Exception as err:
                trial['MAP'] = 0
                print("Eval failed:", trial, "->", err)
            all_rows.append(trial)
        return pd.DataFrame(all_rows)

    coarse_dict = {p: grid_values[p] for p in params_in_use}
    df_coarse = run_grid(coarse_dict)
    best_coarse = df_coarse[df_coarse['MAP'] == df_coarse['MAP'].max()].iloc[0].to_dict()
    print("Coarse optimum found:", best_coarse)

    fine_candidates = {}
    for param in params_in_use:
        c_val = best_coarse[param]
        lower, upper = bounds[param]
        start = max(lower, c_val - 0.5)
        stop = min(upper, c_val + 0.5)
        fine_candidates[param] = [round(v, 2) for v in list(set([lower, upper] +
                                        [start + i * step for i in range(int((stop - start) / step) + 1)])) if lower <= v <= upper]

    df_fine = run_grid(fine_candidates)

    best_fine_row = df_fine[df_fine['MAP'] == df_fine['MAP'].max()]
    best_combination = best_fine_row.iloc[0].to_dict()
    print(f"Fine-tuned best params for {model_name}: {best_combination}")

    return df_fine
```

```python
def display_results(model_name, df):
    param_cols = [c for c in df.columns if c != 'MAP']
    num_params = len(param_cols)

    if num_params == 0:
        print(f"{model_name} has no tunable parameters. Best MAP: {df['MAP'].max():.4f}")
        return

    def style_axes(ax, title, xlabel, ylabel):
        ax.set_title(title, fontsize=14, fontweight='bold')
        ax.set_xlabel(xlabel, fontsize=12)
        ax.set_ylabel(ylabel, fontsize=12)
        ax.grid(True, linestyle='--', alpha=0.5)

    if num_params == 1:
        param = param_cols[0]
        fig, ax = plt.subplots(figsize=(8, 5))
        ax.scatter(df[param], df['MAP'], color='darkorange', edgecolor='k', s=80, alpha=0.8)
        for xi, yi in zip(df[param], df['MAP']):
            ax.text(xi, yi + 0.001, f"{yi:.3f}", ha='center', fontsize=9, fontweight='medium')
        style_axes(ax, f"{model_name}: MAP vs {param}", param, "MAP")
        plt.show()

    elif num_params == 2:
        x_param = 'k' if 'k' in param_cols else param_cols[0]
        y_param = 'b' if 'b' in param_cols else param_cols[1]
        sizes = df['MAP'] * 1200
        colors = df['MAP']

        fig, ax = plt.subplots(figsize=(8, 6))
        scatter = ax.scatter(df[x_param], df[y_param], s=sizes, c=colors, cmap='coolwarm', alpha=0.75, edgecolors='gray')
        cbar = plt.colorbar(scatter, ax=ax)
        cbar.set_label('MAP', fontsize=12)
        for xi, yi, score in zip(df[x_param], df[y_param], df['MAP']):
            ax.text(xi, yi, f"{score:.3f}", ha='center', va='center', fontsize=8, color='white', fontweight='bold')
        style_axes(ax, f"{model_name}: MAP Heat Bubble", x_param, y_param)
        plt.show()

    elif num_params == 3:
        fig = plt.figure(figsize=(10, 7))
        ax = fig.add_subplot(111, projection='3d')
        x, y, z = df['k'].values, df['b'].values, df['sigma'].values
        scores = df['MAP'].values

        scatter = ax.scatter(x, y, z, c=scores, s=scores*120, cmap='plasma', alpha=0.85, edgecolors='k')
        cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
        cbar.set_label('MAP', fontsize=12)
        ax.set_xlabel('k', fontsize=12)
        ax.set_ylabel('b', fontsize=12)
        ax.set_zlabel('sigma', fontsize=12)
        ax.set_title(f"{model_name}: 3D MAP Bubble Plot", fontsize=14, fontweight='bold')
        plt.show()
```

# **تحلیل روش‌های رتبه‌بندی اسناد و تأثیر پارامترها بر دقت بازیابی اطلاعات**

## ۱. **BM25: اصول و نقش پارامترها**

BM25 یکی از رایج‌ترین مدل‌های رتبه‌بندی در بازیابی اطلاعات است که علاوه بر وزن‌دهی ترم‌ها، **طول سند** را نیز در امتیازدهی لحاظ می‌کند. دو پارامتر اصلی در این مدل، \( k \) و \( b \)، نقش حیاتی دارند.

- **پارامتر \( k \) (اشباع فرکانس واژه‌ها):**  
  این پارامتر میزان اثرگذاری تکرار یک واژه در سند را تعیین می‌کند. مقادیر پایین باعث برجسته‌تر شدن تأثیر واژه‌های پرتکرار و مقادیر بالا کاهش این اثر را به دنبال دارند.

- **پارامتر \( b \) (نرمال‌سازی طول سند):**  
  مشخص می‌کند طول سند چه میزان در محاسبه امتیاز لحاظ شود. زمانی که \( b \) نزدیک صفر باشد، طول سند تقریباً نادیده گرفته می‌شود و وقتی به یک نزدیک باشد، اسناد بلندتر نسبت به اسناد کوتاه‌تر جریمه می‌شوند.

**تحلیل عملکرد:**  
مطالعات تجربی نشان داده‌اند که مقادیر میانه مثل \( k = 0.75 \) و \( b = 0.25 \) تعادل مناسبی بین اهمیت فرکانس واژه‌ها و طول سند برقرار کرده و **MAP** بهینه‌ای را ارائه می‌کنند.


## ۲. **روش ۱: مجموع وزن‌های IDF بدون در نظر گرفتن TF و طول سند**

این روش ساده‌ترین مدل امتیازدهی است که تنها **وزن IDF ترم‌های مشترک بین پرسش و سند** را جمع می‌کند.  

### **ویژگی‌ها و محدودیت‌ها:**
- تکرار واژه‌ها (TF) و طول سند در محاسبات لحاظ نمی‌شوند.  
- سندهای بلند و کوتاه به یک شکل ارزیابی می‌شوند و تفاوت بین اسنادی که واژه‌ها را چند بار استفاده کرده‌اند یا یک بار، مشخص نمی‌شود.

### **پیامدها:**
- نادیده گرفتن اطلاعات آماری کلیدی باعث کاهش دقت مدل می‌شود.  
- MAP نسبت به مدل‌های پیشرفته‌تر مانند BM25 پایین‌تر است.



## ۳. **روش ۲: کنترل اثر اشباع فرکانس واژه‌ها با k**

این روش از فرمول مشابه BM25 استفاده می‌کند، اما **تنها پارامتر قابل تنظیم k** است و طول سند همچنان نادیده گرفته می‌شود.

### **تحلیل:**
- کاهش k از 1.5 به 0.5 باعث افزایش تدریجی MAP شده است.  
- مقدار بهینه \( k = 0.5 \) MAP را به 0.1254 رسانده است.  
- این نتیجه نشان می‌دهد که تأکید بیشتر بر واژه‌های پرتکرار می‌تواند نقش مهمی در افزایش دقت بازیابی داشته باشد.


## ۴. **روش ۳: ترکیب وزن واژه‌ها و طول سند**

در این روش، علاوه بر وزن ترم‌ها، طول سند نیز در فرمول لحاظ می‌شود. پارامتر k تعادل میان تأثیر TF و طول سند را مشخص می‌کند.

### **یافته‌ها:**
- کاهش تدریجی k باعث افزایش MAP شده است.  
- مقدار بهینه \( k = 0.5 \) MAP برابر با 0.1936 ارائه کرده است.  
- این نشان می‌دهد که در این مجموعه داده، اهمیت واژه‌های پرتکرار نسبت به طول سند نقش مؤثرتری داشته است.

## ۶. **روش ۵: تأکید بر واژه‌های نادر (IDF^2)**

```python
methods = {
    'BM25': rm.method6_score,
    'Method 1': rm.method1_score,
    'Method 2': rm.method2_score,
    'Method 3': rm.method3_score,
    'Method 4': rm.method4_score,
    'Method 5': rm.method5_score,
    'Method 6': rm.method6_score,
    'Method 7': rm.method7_score,
    'Pivoted Length V1': rm.pivoted_length_v1_score,
    'Pivoted Length V2': rm.pivoted_length_v2_score
}

best_params_by_method = {
    'BM25': {'k': 0.75, 'b': 0.25},
    'Method 1': {},
    'Method 2': {'k': 0.5},
    'Method 3': {'k': 0.5},
    'Method 4': {},
    'Method 5': {'k': 1, 'b': 0},
    'Method 6': {'k': 1.25, 'b': 0.5, 'delta': 1.0},
    'Method 7': {'k': 1.5, 'b': 0.75},
    'Pivoted Length V1': {'b': 0.75},
    'Pivoted Length V2': {'b': 0.75}
}
```

```python
def compute_all_ap(queries, relevance_judgments, inverted_index, models, best_params):
    def average_precision(retrieved, relevant_dict, min_score=3):
        relevant_set = {doc for doc, score in relevant_dict.items() if score >= min_score}
        if not relevant_set:
            return 0.0
        num_relevant = 0
        prec_sum = 0.0
        for rank, doc_id in enumerate(retrieved, 1):
            if doc_id in relevant_set:
                num_relevant += 1
                prec_sum += num_relevant / rank
        return prec_sum / len(relevant_set)

    all_ap_results = {}
    
    for q in queries:
        qid = q.query_id
        relevant_docs = relevance_judgments.get(qid, {})

        all_ap_results[qid] = {}

        for name, func in models.items():
            params = best_params.get(name, {})
            try:
                scores = func(q, inverted_index, **params)
            except TypeError:
                scores = func(q, inverted_index)
            retrieved_sorted = sorted(scores.items(), key=lambda x: x[1], reverse=True)
            retrieved_ids = [doc for doc, _ in retrieved_sorted]
            all_ap_results[qid][name] = average_precision(retrieved_ids, relevant_docs)

    return all_ap_results

all_ap_results = compute_all_ap(
    queries=queries,
    relevance_judgments=relevance_judgments,
    inverted_index=inverted_index,
    models=models,
    best_params=best_params_by_model
)

df_ap_results = pd.DataFrame([
    {**{'Query': qid}, **scores} 
    for qid, scores in all_ap_results.items()
])

print("\nDataframe version structure:")
print(f"Number of queries: {len(all_ap_results)}")
print(f"Sample query keys: {list(all_ap_results.keys())[:5]}")
print("DataFrame version:")
df_ap_results.head()
```

```python
def compare_models_to_bm25(all_ap_results):
    query_ids = list(all_ap_results.keys())
    bm25_scores = [all_ap_results[q]['BM25'] for q in query_ids]
    model_keys = [k for k in next(iter(all_ap_results.values())).keys() if k != 'BM25']

    better_counts = dict.fromkeys(model_keys, 0)
    worse_counts = dict.fromkeys(model_keys, 0)
    p_vals = {}

    def count_better_worse(model_scores, bm25_scores):
        better, worse = 0, 0
        for m, b in zip(model_scores, bm25_scores):
            if m > b:
                better += 1
            elif m < b:
                worse += 1
        return better, worse

    for model in model_keys:
        model_scores = [all_ap_results[q][model] for q in query_ids]
        t_stat, p_value = ttest_rel(model_scores, bm25_scores)
        p_vals[model] = p_value
        better_counts[model], worse_counts[model] = count_better_worse(model_scores, bm25_scores)

    for model in model_keys:
        print(f"{model}:")
        print(f"  Number of queries with better performance than BM25: {better_counts[model]}")
        print(f"  Number of queries with worse performance than BM25: {worse_counts[model]}")
        print(f"  p-value: {p_vals[model]}")
        significance = "statistically significant" if p_vals[model] < 0.05 else "not statistically significant"
        print(f"  The difference in performance between {model} and BM25 is {significance}.\n")

compare_models_to_bm25(all_ap_results)
```

```python
def evaluate_methods(inverted_index, methods_dict, best_params_dict, queries, relevance_judgments):
    def compute_ap_for_method(method_fn, param_dict):
        ap_values = []
        for q in queries:
            res = ev.evaluate_scoring_method(
                inverted_index, method_fn, [q], relevance_judgments, params=param_dict
            )
            ap_values.append(res['MAP'])
        return ap_values

    baseline_name = list(methods_dict.keys())[0]
    baseline_fn = methods_dict[baseline_name]
    baseline_params = best_params_dict[baseline_name]
    baseline_ap = compute_ap_for_method(baseline_fn, baseline_params)

    all_results = []

    for idx, (name, fn) in enumerate(methods_dict.items()):
        params = best_params_dict[name]
        method_ap = compute_ap_for_method(fn, params)
        overall_metrics = ev.evaluate_scoring_method(inverted_index, fn, queries, relevance_judgments, params=params)

        if idx == 0:
            better, worse, pval_str = '-', '-', '-'
        else:
            better = sum(ma > ba for ma, ba in zip(method_ap, baseline_ap))
            worse = sum(ma < ba for ma, ba in zip(method_ap, baseline_ap))
            _, raw_pval = ttest_rel(method_ap, baseline_ap)
            pval_str = f"{raw_pval:.2e}"

        metrics_row = {
            'Method': name,
            'MAP': overall_metrics.get('MAP', 0),
            'MRR': overall_metrics.get('MRR', 0),
            'Precision@3': overall_metrics.get('Precision@3', 0),
            'Precision@10': overall_metrics.get('Precision@10', 0),
            'Recall@3': overall_metrics.get('Recall@3', 0),
            'Recall@10': overall_metrics.get('Recall@10', 0),
            'nDCG@10': overall_metrics.get('nDCG@10', 0),
            'Better': better,
            'Worse': worse,
            'p-val (MAP)': pval_str
        }

        all_results.append(metrics_row)

    return pd.DataFrame(all_results)
```

```python
df_evaluation = evaluate_methods(
    inverted_index=inverted_index,
    methods_dict=methods,
    best_params_dict=best_params_by_method,
    queries=queries,
    relevance_judgments=relevance_judgments
)

df_evaluation
```
